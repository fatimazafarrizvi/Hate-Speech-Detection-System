```python
import pandas as pd

# Load your main dataset
main_df = pd.read_csv("merged_hate_speech_dataset.csv")

# Load Davidson dataset
davidson_df = pd.read_csv("labeled_data.csv.zip")

# Extract only Offensive Language and Neither
non_hate_df = davidson_df[davidson_df["class"].isin([1, 2])].copy()

# Keep only the tweet column
non_hate_df = non_hate_df[["tweet"]]

# Rename to match main dataset
non_hate_df.rename(columns={"tweet": "text"}, inplace=True)

# Assign Non-Hate label
non_hate_df["label"] = 0

# Merge with the main dataset
final_df = pd.concat([main_df, non_hate_df], ignore_index=True)

# Remove missing values
final_df.dropna(subset=["text"], inplace=True)

# Remove duplicate texts
final_df.drop_duplicates(subset=["text"], inplace=True)

# Shuffle the dataset
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save
final_df.to_csv("final_hate_speech_dataset.csv", index=False)

# Display statistics
print("Final Dataset Shape:", final_df.shape)
print("\nLabel Distribution:")
print(final_df["label"].value_counts())
```

```python
# ============================================
# Hate Speech Dataset Cleaning Pipeline
# Run this in Colab after uploading your CSV
# ============================================

import pandas as pd
import re

# --- 1. Load ---
df = pd.read_csv('final_merged_hate_speech_dataset.csv')
print(f"Original shape: {df.shape}")

# --- 2. Drop missing labels ---
df = df.dropna(subset=['label', 'text']).copy()
df['label'] = df['label'].astype(int)  # 0.0/1.0 -> 0/1
print(f"After dropping NaNs: {df.shape}")

# --- 3. Normalize social media artifacts (consistent preprocessing) ---
def normalize_text(text):
    text = str(text)

    # Unify placeholder style: turn raw mentions into <user>
    text = re.sub(r'@\w+', '<user>', text)

    # Turn URLs into <url>
    text = re.sub(r'http\S+|www\.\S+', '<url>', text)

    # Strip leading "RT" retweet marker
    text = re.sub(r'^RT\s*[:]?\s*', '', text, flags=re.IGNORECASE)

    # Collapse repeated whitespace/newlines
    text = re.sub(r'\s+', ' ', text).strip()

    return text

df['text_clean'] = df['text'].apply(normalize_text)

# --- 4. Flag decorative/spam junk (heavy non-ASCII ratio, low real-word content) ---
def junk_score(text):
    if len(text) == 0:
        return 1.0
    non_ascii = sum(1 for c in text if ord(c) > 127)
    return non_ascii / len(text)

df['non_ascii_ratio'] = df['text_clean'].apply(junk_score)
junk_mask = df['non_ascii_ratio'] > 0.3  # tune threshold as needed
print(f"\nFlagged as likely junk/spam (>30% non-ASCII): {junk_mask.sum()} rows")
df = df[~junk_mask].copy()

# --- 5. Deduplicate (exact + case-insensitive near-duplicates) ---
before = len(df)
df['text_norm'] = df['text_clean'].str.lower().str.strip()
df = df.drop_duplicates(subset=['text_norm'], keep='first').copy()
print(f"Removed {before - len(df)} duplicate/near-duplicate rows")

# --- 6. Handle very short texts (decide: keep, flag, or drop) ---
df['word_count'] = df['text_clean'].str.split().apply(len)
very_short = df['word_count'] <= 2
print(f"\nVery short texts (<=2 words): {very_short.sum()}")
print("Sample:", df[very_short][['text_clean', 'label']].head(5).to_dict('records'))

# OPTION A: keep them but flag for review
df['is_very_short'] = very_short

# OPTION B: drop them entirely — uncomment if you decide context-free
# single words/slurs aren't useful for your use case
# df = df[~very_short].copy()

# --- 7. Final cleanup: drop helper columns, keep clean text as primary ---
df_final = df[['text_clean', 'label']].rename(columns={'text_clean': 'text'})
df_final = df_final.reset_index(drop=True)

# --- 8. Stats before/after ---
print(f"\n{'='*50}")
print(f"FINAL DATASET: {df_final.shape}")
print(f"Class balance:\n{df_final['label'].value_counts(normalize=True)}")
print(f"{'='*50}")

# --- 9. Save ---
df_final.to_csv('hate_speech_cleaned.csv', index=False)
print("\nSaved as hate_speech_cleaned.csv")

# Optional: download in Colab
from google.colab import files
files.download('hate_speech_cleaned.csv')
```
